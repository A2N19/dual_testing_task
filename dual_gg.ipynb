```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_kingfisher():
    base_urls = [
        "https://kingfisher.kz/ryba/",
        "https://kingfisher.kz/ikra/",
        "https://kingfisher.kz/myaso/",
        "https://kingfisher.kz/antipasti/",
        "https://kingfisher.kz/bakaleya/",
        "https://kingfisher.kz/sneki-i-orehi/",
        "https://kingfisher.kz/maslo/",
        "https://kingfisher.kz/cheeses/",
    ]

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"
    }

    data = []

    for base_url in base_urls:
      
        response = requests.get(base_url, headers=headers)
        if response.status_code != 200:
            print(f"gg полное: {base_url}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')

      
        category = soup.find("h1", class_="title").text.strip()

        
        products = soup.find_all("div", class_="goodsBlock")
        for product in products:
            try:
                product_name = product.find("a", class_="title").find("span").text.strip()

                
                product_price = product.find("span", class_="price").find("span", class_="new").text.strip()

              
                city = "Астана"  

                data.append({
                    "Название продукта": product_name,
                    "Категория продукта": category,
                    "Цена продукта": product_price,
                    "Город": city
                })
            except AttributeError:
                print(f"gg полное: {base_url}")

    with open("kingfisher_data.csv", "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["Название продукта", "Категория продукта", "Цена продукта", "Город"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerows(data)

    print("Данные успешно сохранены в kingfisher_data.csv")

if __name__ == "__main__":
    scrape_kingfisher()
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_kingfisher():
    base_urls = [
        "https://kingfisher.kz/ryba/",
        "https://kingfisher.kz/ikra/",
        "https://kingfisher.kz/myaso/",
        "https://kingfisher.kz/antipasti/",
        "https://kingfisher.kz/bakaleya/",
        "https://kingfisher.kz/sneki-i-orehi/",
        "https://kingfisher.kz/maslo/",
        "https://kingfisher.kz/cheeses/",
    ]

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"
    }

    data = []

    for base_url in base_urls:
        response = requests.get(base_url, headers=headers)
        if response.status_code != 200:
            print(f"gg полное: {base_url}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')

        category = soup.find("h1", class_="title").text.strip()

        # Ищем блоки с товарами
        products = soup.find_all("div", class_="goodsBlock")
        for product in products:
            try:
                
                product_name = product.find("a", class_="title").find("span").text.strip()

                
                product_price = product.find("span", class_="price").find("span", class_="new").text.strip()

                
                city = "Алматы"  

                # Добавляем данные в список
                data.append({
                    "Название продукта": product_name,
                    "Категория продукта": category,
                    "Цена продукта": product_price,
                    "Город": city
                })
            except AttributeError:
                print(f"gg полное: {base_url}")

    # Сохраняем данные в CSV
    with open("kingfisher_data2.csv", "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["Название продукта", "Категория продукта", "Цена продукта", "Город"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerows(data)

    print("Данные успешно сохранены в kingfisher_data.csv")

if __name__ == "__main__":
    scrape_kingfisher()
```
